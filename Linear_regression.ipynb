```python
import numpy as np
import pandas as pd
```

Create a Dummy Relational DataBase for Salary Prediction

```python
# USe Dictionary for creating Data or Load a CSV file which has relational data
data={'Name':['Nandu','Anupriya','Ashna','digil'],
      'Age':[26,24,24,26],
      'Job':['Intern','Engineer','Engineer','Senior Engineer'],
      'Salary':[None,500000,500000,600000]}
data
```

Pandas For Saving the data into a dataframe

```python
df=pd.DataFrame(data)
df
```

Use Fillna to remove any "None" Values

```python
df['Salary'] = df['Salary'].fillna(1)
```

```python
df
```

```python
# X = Features and y= target variable
# here I am using full data as X for training, no validation and test dataset
X = df[['Age', 'Job']]
y = df['Salary']
```

Machine lEarning model libraries

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

```python
# Preprocessing pipeline (OneHotEncoder for 'Job' column)
preprocessor = ColumnTransformer(
    transformers=[
        ('job', OneHotEncoder(), ['Job'])
    ], remainder='passthrough')
```

```python
# Define the linear regression model pipeline
model = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('regressor', LinearRegression())
])
```

```python
# Train the model
model.fit(X, y)
```

```python

new_employee = pd.DataFrame({
    'Age': [22],
    'Job': ['Engineer']
})
```

```python

new_employee = pd.DataFrame({
    'Age': [28],
    'Job': ['Senior Engineer']
})
```

```python

new_employee = pd.DataFrame({
    'Age': [26],
    'Job': ['Intern']
})
```

```python
predicted_salary = model.predict(new_employee)
print(f"Predicted salary: {predicted_salary[0]}")
```
